### Seeing all the postlinks

### PostLinks
* Id primary key
* CreationDate when the link was created
* PostId id of source post
* RelatedPostId id of target/related post
* LinkTypeId type of link
* ** 1 = Linked (PostId contains a link to RelatedPostId)
* ** 3 = Duplicate (PostId is a duplicate of RelatedPostId)


```python
dupes_dict=dict()
```

```python
total_links=0
weird_links=0
threshold=0
for i in range(0, 7+1):
    with open(f"./data_files/Postlinks_result/post_{i}.json",'r') as fd:
        df=json.load(fd)
    for curr_key, curr_val in df.items():
        if curr_val['LinkTypeId']!='3':
            continue
        yr_of_link=int(curr_val['CreationDate'].split("-")[0])
        if yr_of_link>2013:
            continue
        #print(yr_of_link)
        n1=curr_val['PostId']
        n2=curr_val['RelatedPostId']
        if (n1 not in creation_dates) or (n2 not in creation_dates) or (n1==n2):
            continue
        #print(f"{n1} -> {n2}")
        child_node=n1
        parent_node=n2
        total_links+=1

        if creation_dates[parent_node]>creation_dates[child_node]:
            print("Weird")
            print(parent_node)
            print(child_node)
            weird_links+=1
            continue
        for the_node in [child_node, parent_node]:
            if the_node not in dupes_dict:
                dupes_dict[the_node]={"q_id":the_node, "creation_date":creation_dates[the_node],"parent_q_list":[],"child_q_list":[]}
        dupes_dict[child_node]['parent_q_list'].append(parent_node)
        dupes_dict[parent_node]['child_q_list'].append(child_node)
     
        #break
```
