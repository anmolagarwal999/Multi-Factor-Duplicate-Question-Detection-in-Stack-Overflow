## Run loop

```python
PREFIX_FILE_PATH="../data_files/cleaned_pp_result"
```

```python
# stores tuples of form <time,post id, file num>
details_arr=[]
```

```python
lb_file_idx=0
ub_file_idx=102

for file_id in range(lb_file_idx, ub_file_idx+1):
    with open(PREFIX_FILE_PATH+f"/post_{file_id}.json", 'r') as fd:
        df=json.load(fd)
    print(f"{file_id} -> {len(df.keys())}")
    for curr_key, curr_val in df.items():
        curr_date=curr_val["CreationDate"]
        details_arr.append((curr_date, curr_key, file_id))
    
```

```python
len(details_arr)
```

```python
prev=None
cnt=0
for curr_elem in details_arr:
    now_date=curr_elem[0]
    
    if prev==None or now_date>=prev[0]:
        # valid case
        pass
    else:
        cnt+=1
        print(f"{prev}: {curr_elem}")
    prev=curr_elem
        
```
